```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

iloc example:\
first_five = data.iloc[:5,:-1]\
first arg represent the rows and second arg represent the columns\
:5 means first 5 rows and :-1 means all columns except the last one


```python
data = pd.read_csv("dataset/eeg_with_numeric_labels.csv")
first_five = data.iloc[:5,:-1]
first_five
```

time domain vis

```python
# Time domain visualization
plt.figure(figsize=(12, 6))

plt.plot(first_five.iloc[3, :], label=f'Score {data.iloc[3]["eeg_251"]}')
plt.title('EEG Signals - Time Domain')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Frequency domain visualization
plt.figure(figsize=(12, 6))

spectrum = np.fft.fft(first_five.iloc[3, :])

freq = np.fft.fftfreq(len(spectrum))
plt.plot(freq, np.abs(spectrum), label=f'Record {data.iloc[3].eeg_251}')
plt.title('EEG Signals - Frequency Domain')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.legend()
plt.grid(True)
#plt.xlim(0, 125)  # Limit the x-axis to show only positive frequencies 
plt.show()
```
